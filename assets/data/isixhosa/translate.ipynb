```python
import pandas as pd
import numpy as np
```

```python
translations = pd.read_csv('translations.tsv', sep='	', names=["filename", "translation"])
```

```python
translations
```

```python
folder = r'sounds_wav/'
```

```python
import re
import os

for index, (value) in translations.iterrows():
    old_file_name = value['filename']
    translation = value['translation']
    new_file_name = ' '.join(re.split('[^a-zA-Z]', translation))
    new_file_name = new_file_name.lower()
    new_file_name = re.sub(' +', '_', new_file_name)
    new_file_name = re.sub('_$', '', new_file_name)
    new_file_name = new_file_name + '.wav'
    
    print(folder + old_file_name)
    #os.rename(folder + )
```
